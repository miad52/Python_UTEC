## **Preparación de los Datos**

```python
df.info()
```

```python
df2 = df[["id","host_id", "listing_url","host_name", "host_is_superhost", "name", "neighbourhood", "neighbourhood_cleansed", "host_response_rate", "latitude", "longitude", "room_type", "price", "number_of_reviews", "review_scores_rating"  ]]
```

```python
df2.head()
```

```python
#Limpiamos algunas variables como price y host_response_rate
def limpiar(a):
  return str(a).replace("$","").replace(",","")
df2["price"] = df2["price"].apply(limpiar)#replace("[$,]", "")
df2["host_response_rate"] = df2["host_response_rate"].str.replace("%","")
```

```python
#Convertimos el precio y host_response_rate a float
df2['price'] = df2['price'].astype(float)
df2["host_response_rate"] = df2['host_response_rate'].astype(float)
```

## **Hay mucho outlier, entonces tenemos que hacer alguna limpieza**

- vemos que es lo que esta pasando

```python
df2[df2['price']>10000].head()
```

```python
#Verificamos que si hay valores atípicos por debajo de 10, 50, 100
df2[df2['price']<50].head()
```

```python
#vamos a eliminar los que estan por arriba de los 10k reales
df2.drop(df2[df2['price']>10000].index,inplace=True)
```

```python
#Generamos nuevamente la imagen
sns.boxplot(data = df2, y= "price")
```

```python
df_filter = df2[(df2['price']<100) & (df2['room_type']=='Entire home/apt')] # Son outliers? Entire home/apt  Private room
df_filter
```

```python
# Importar librería Folium
import folium
```

```python
#Crea Mapa
map = folium.Map(location=[-22.950463, -43.165169], zoom_start=12)
```

```python
map
```

```python
set(df2["host_is_superhost"])
```

```python
df3 = df2.loc[df2['host_is_superhost'] == 't']
df3.head()
```

```python
from folium import plugins
from folium.plugins import FastMarkerCluster

df4 = df3#df2[df2['room_type']=='Private room']
locations = list(zip(df4.latitude, df4.longitude,df4.id))

FastMarkerCluster(data=locations).add_to(map)
map
```

```python
listTuplas = zip(df3['latitude'],df3['longitude'])

#Crear marcadores sobre el mapa en base a latitude y longitude
for lat, lon in listTuplas:
  marker = folium.Marker(location=[lat, lon])
  marker.add_to(map)
map
```
